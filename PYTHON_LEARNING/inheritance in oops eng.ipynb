```python
class Parent:
    def getParent(self):
        return "Parent"
```

```python
class Child(Parent):
    def getChild(self):
        return "Child"
```

```python
class GrandChild(Child):
    def getGrandChild(self):
        return "GrandChild"
```

```python
class GreatGrandChild1(GrandChild):
    def getGreatGrandChild1(self):
        return "GreatGrandChild1"
class GreatGrandChild2(GrandChild):
    def getGreatGrandChild2(self):
        return "GreatGrandChild2"
```

```python
parent = Parent()
```

```python
parent.getParent()
```

```python
grand_child = GrandChild()
```

```python
grand_child.getParent()
```

```python
grand_child.getGrandChild()
```

```python
great_grand_child1 = GreatGrandChild1()
great_grand_child2 = GreatGrandChild2()
```

```python
great_grand_child1.getGreatGrandChild1()
```

```python
class level1:
    def getLevel1(self):
        return "level1"
class level2(level1):
    def getLevel2(self):
        return "level2"
class level3(level2):
    def getLevel3(self):
        return "level3"
class level4(level3):
    def getLevel4(self):
        return "level4"
class level5(level3):
    def getLevel5(self):
        return "level5"
class level6(level4,level5):
    def getLevel6(self):
        return "level6"
```
