```python
class Encapsulate:
    def __init__(self):
        self.__a = 0
    def set_a(self, a):
        self.__a = a
    def get_a(self):
        return self.__a
```

```python
encap = Encapsulate()
```

```python
encap.set_a(1)
```

```python
encap.get_a()
```

```python
encap.a
```

```python
encap.__a
```

```python
encap._Encapsulate__a
```

```python
encap._Encapsulate__a = 2
```

```python
encap.get_a()
```

```python
class bank_acount:
    
    def __init__(self , balance ):
        self.__balance = balance
    
    def deposit(self , amount ) :
        self.__balance = self.__balance + amount
    
    def withdraw(self , amount) : 
        if self.__balance >= amount : 
            self.__balance = self.__balance -amount
            return True
        else : 
            return False
        
    def get_balance(self) : 
        return self.__balance           
```

```python
obj_bank_account = bank_acount(1000)
```

```python
obj_bank_account.get_balance()
```

```python
obj_bank_account.deposit(6000)
```

```python
obj_bank_account.get_balance()
```

```python
obj_bank_account.withdraw(10000)
```

```python
obj_bank_account.withdraw(2000)
```

```python
obj_bank_account.get_balance()
```

```python
class Bank:
    def __init__(self,amount):
        self.amount = amount
    def take(self,money):
        if(money<=self.amount):
            self.amount -= money
            return money
        return -1
    def put(self,money):
        if(money>0):
            self.amount += money
            return self.amount
        return -1
    def checkAmount(self):
        return self.amount


    
```

```python
bank = Bank(100)
```

```python
bank.checkAmount()
```

```python
bank.checkAmount
```

```python
bank.take(100)
```

```python
bank.checkAmount()
```

```python
bank.put(100)
```

```python
bank.checkAmount()
```

```python
bank.take(101)
```

```python
bank.put(-100)
```
