```python
import matplotlib.pyplot as plt

# Data for pie chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]
colors = ['red', 'yellow', 'pink', 'brown']
explode = (0.1, 0, 0, 0)  # explode the first slice

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    textprops={'fontsize': 14, 'color': 'black', 'fontweight': 'bold'}
)
plt.title("Fruit Distribution", fontsize=16, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()

```

```python
import numpy as np

# Data for bar chart
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [2500, 4000, 1500, 3000, 2000]

# Calculate total and average
total_revenue = sum(sales)
average_revenue = total_revenue / len(sales)
below_avg_indices = [i for i, val in enumerate(sales) if val < average_revenue]
max_sale_index = sales.index(max(sales))

# Bar colors and borders
bar_colors = ['green' if i != max_sale_index else 'orange' for i in range(len(sales))]
edge_colors = ['black'] * len(sales)

# Create bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(products, sales, color=bar_colors, edgecolor=edge_colors)

# Labeling
plt.title("Product Sales", fontsize=18, fontweight='bold')
plt.xlabel("Products", fontsize=14)
plt.ylabel("Sales ($)", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Annotate bars
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 100,
             f"${sales[i]}", ha='center', fontsize=12)

# Highlight below average sales
for i in below_avg_indices:
    plt.text(i, sales[i] - 300, "Below Avg", ha='center', fontsize=10, color='red')

plt.tight_layout()
plt.show()

```
