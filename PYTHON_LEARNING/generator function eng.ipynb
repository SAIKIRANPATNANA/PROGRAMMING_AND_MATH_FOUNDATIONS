```python
def test_fib(n) :
    a,b = 0,1
    for i in range(n):
        yield a 
        a,b = b ,a+b 
```

```python
def testFib(n):
    a,b = 0,1 
    for i in range(n):
        yield a
        a , b = b, a+b
```

```python
def countDown(n):
    while(n>0):
        yield n
        n -= 1
```

```python
for num in countDown(10):
    print(num)
```

```python
def infiniteNumbers():
     n = 1
     while(True):
        yield n
        n += 1
```

```python
for num in infiniteNumbers():
    if num>10:
        break
    print(num)
```
