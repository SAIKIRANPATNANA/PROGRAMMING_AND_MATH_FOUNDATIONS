### Abstraction (Hiding unnecessary details, showing only essential features)(Hiding unnecessary details, showing only essential features)

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def drive(self):
        pass

class Car(Vehicle):
    def drive(self):
        print("Car is being driven")

class Bike(Vehicle):
    def drive(self):
        print("Bike is being ridden")

# Using abstraction
v1 = Car()
v2 = Bike()

v1.drive()
v2.drive()
```

 ### Encapsulation (Protecting data by restricting access)

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")

    def get_balance(self):
        return self.__balance

# Using encapsulation
account = BankAccount(1000)
account.deposit(500)
print("Balance:", account.get_balance())
print(account._BankAccount__balance)
# Trying to access the private variable directly (will throw an error)
# print(account.__balance)  # Uncomment to see the error
```

```python
##Polymorphism:
class Polymorphism1:
    def drive(self):
        return 'car'
class Polymorphism2:
    def drive(self):
        return 'bike'
for obj in [Polymorphism1(), Polymorphism2()]:
    print(obj.drive())
##Encapsulation:
class Encapsulation:
    def __init__(self, amount):
        self.__amount = amount
    def deposit(self, amount):
        self.__amount += amount
    def withdraw(self, amount):
         if(self.__amount>=amount):
                self.__amount -= amount
                return self.__amount
         return -1
encapsulation = Encapsulation(100)
encapsulation.deposit(20)
encapsulation.withdraw(130)
```

```python
class Animal:
    # Parent class method
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    # Overriding the method in the child class
    def sound(self):
        print("Bark!")

class Cat(Animal):
    # Overriding the method in the child class
    def sound(self):
        print("Meow!")

# Creating objects
dog = Dog()
cat = Cat()

dog.sound()  # Output: Bark! (overridden in Dog class)
cat.sound()  # Output: Meow! (overridden in Cat class)
```
