```python
class data_science:
    def syllabus(self) : 
        print("this is my method for data science syllbaus " )
```

```python
class web_dev :
    def syllabus(self) : 
        print("this is my method for web dev " )
```

```python
def class_parcer(class_obj) : 
    for i in class_obj : 
        i.syllabus()
```

```python
obj_data_science= data_science()
```

```python
obj_web_dev = web_dev()
```

```python
class_ojb = [obj_data_science , obj_web_dev]
```

```python
class_parcer(class_ojb)
```

```python
class Mama:
    def common(self):
        print("Common Mama")
class Bro:
    def common(self):
        print("Common Bro")
mama = Mama()
bro = Bro()
def track(objs):
    for obj in objs:
        obj.common()
objs = [mama, bro]
track(objs)
```
