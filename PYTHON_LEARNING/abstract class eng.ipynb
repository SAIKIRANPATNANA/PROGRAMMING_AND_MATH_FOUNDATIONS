```python
# import abc
class PWSkills:
    # @abc.abstractmethod 
    def getStuff(self):
        pass
    # @abc.abstractmethod
    def getExtraStuff(self):
        pass
    # @abc.abstractmethod
    def getDoubleExtraStuff(self):
        pass
```
