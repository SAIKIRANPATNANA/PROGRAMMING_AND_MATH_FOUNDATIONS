## RE BASED WEB SCRAPPING 

```python
# Importing the required libraries.
import requests
from bs4 import BeautifulSoup
import re
from pprint import pprint
# Requesting the HTML from the target website.
url = "https://sandbox.oxylabs.io/products"
page = requests.get(url)

# Selecting data.
soup = BeautifulSoup(page.content, "html.parser")
products = soup.find_all("div", class_="product-card")

# Processing data using Regular Expressions.
re_titles = r'class="title css-7u5e79 eag3qlw7">(.*?)<\/h4>'
re_prices = r'class="price-wrapper css-li4v8k eag3qlw4">(.*?)<\/div>'

data = []

for product in products:
   product_html = str(product)
   title = re.findall(re_titles, product_html)
   price = re.findall(re_prices, product_html)
   data.append((title, price))

print(data)
https://sandbox.oxylabs.io/products
# Saving the output.
with open("output.txt", "w", encoding="utf-8") as f:
   for title, price in data:
       f.write(f"{title}\t{price}\n")
```

```python
import requests
from bs4 import BeautifulSoup

# Amazon product URL (Replace with actual product URL)
product_url = 'https://www.amazon.in/s?k=gate+cse+pyq'

# Send GET request to the URL
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
response = requests.get(product_url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the product details using the provided CSS selectors
try:
    # Product name
    name = soup.select_one('#\\32 795bdfd-2683-4bd8-b563-5ad22ff7ce91 .s-title-instructions-style h2 span').get_text(strip=True)

    # Rating
    rating = soup.select_one('#\\32 795bdfd-2683-4bd8-b563-5ad22ff7ce91 .a-icon.a-icon-star-small span').get_text(strip=True)

    # Number of ratings
    num_of_ratings = soup.select_one('#\\32 795bdfd-2683-4bd8-b563-5ad22ff7ce91 .a-section.a-spacing-none.a-spacing-top-micro span a span').get_text(strip=True)

    # Price
    price = soup.select_one('#\\32 795bdfd-2683-4bd8-b563-5ad22ff7ce91 .a-price-whole').get_text(strip=True)

    # Company/Brand
    company = soup.select_one('#\\32 795bdfd-2683-4bd8-b563-5ad22ff7ce91 .s-title-instructions-style .a-section span:nth-of-type(2)').get_text(strip=True)

    # Print the extracted details
    print(f"Product Name: {name}")
    print(f"Rating: {rating}")
    print(f"Number of Ratings: {num_of_ratings}")
    print(f"Price: {price}")
    print(f"Company/Brand: {company}")

except AttributeError as e:
    print(f"An error occurred: {e}")
```
