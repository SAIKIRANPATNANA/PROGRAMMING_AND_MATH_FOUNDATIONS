```python
import re
```

```python
s = 'A computer science portal for  CSE students'

match = re.search(r'portal', s)
print('Match Group:', match.group())
print('Start Index:', match.start())
print('End Index:', match.end())
```

```python
str = """Hello my Number is 123456789 and
            my friend's number is 987654321"""
```

```python
re.findall('\d+', str)
```

```python
p = re.compile(r'[a-e]')
p.findall("Aye, said Mr. Gibenson Stark")
```

```python
p = re.compile(r'ab*')
p.findall('ababbaabbb')
```

```python
print(re.split('\W+', 'Words, words , Words', maxsplit=0))
print(re.split('\W+', "Word's words Words", maxsplit=2))
print(re.split('\W+', "Word's words Words", maxsplit=3))
print(re.split('\W+', 'On 12th Jan 2016, at 11:02 AM'))
print(re.split('\d+', 'On 12th Jan 2016, at 11:02 AM'))
```

```python
print(re.sub('ub', '~*', 'Subject has Uber booked already', 
             flags=re.IGNORECASE))
print(re.sub('ub', '~*', 'Subject has Uber booked already'))
print(re.sub('ub', '~*', 'Subject has Uber booked already',
             count=1, flags=re.IGNORECASE))
print(re.sub(r'\sAND\s', ' & ', 'Baked Beans And Spam', 
             flags=re.IGNORECASE))
```
