```python
def deco(func):
    def inner_deco():
        print("this is the start of my fun")
        func()
        print("this is the end of my fun")
    return inner_deco
        
```

```python
@deco
def test1():
    print(4+5)
```

```python
test1()
```

```python
def prasanna(sai):
    def inner_prasanna():
        print('Hey Prasanna')
        sai()
        print('Oh Prasanna')
    return inner_prasanna
```

```python
@prasanna
def sai():
    print('Love You Prasanna')
```

```python
sai()
```

```python
import time
def timer_test(func) :
    def timer_test_inner():
        start = time.time()
        func()
        end = time.time()
        print(end-start)
    return timer_test_inner
```

```python
@timer_test
def test2():
    print(45+67)
```

```python
def TimeCounter(loop):
    def innerTimeCounter():
        t1 = time.time()
        loop()
        t2 = time.time()
        print(t2-t1)
    return innerTimeCounter
```

```python
@TimeCounter
def loop():
    for i in range(1,10000000):
        pass
```

```python
loop()
```
