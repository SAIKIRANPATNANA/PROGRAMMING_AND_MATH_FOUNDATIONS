```python
class pwskills1:
    def student_details(self , name , mail_id , number) : 
        print(name , mail_id, number)
        
    @staticmethod
    def mentor_mail_id(mail_id):
        print(mail_id)
        
    @staticmethod
    def mentor_class(list_mentor) : 
        print(list_mentor)
        pwskills1.mentor_mail_id(["krish@gmail.com" , "sudh@gmail.com"])
        
    @classmethod
    def class_name(cls,class_name):
        cls.mentor_class(["sudh" , "krish"])
        
    def mentor(self , mentor_list) :
        print(mentor_list)
        self.mentor_class(["krish" , "sudh"])
```

```python
class PWSkills:
    def CollectStudentStuff(self, name, mail, number):
        print(name, mail, number)
    @staticmethod 
    def WishPrasanna(msg):
        print(msg)
    def greetPrasanna(self,msg):
        print(msg)
    @classmethod
    def HeyPrasanna(cls,msg):
        cls.WishPrasanna(msg)
    def WohPrasanna(self,msg):
        self.greetPrasanna(msg)
```
