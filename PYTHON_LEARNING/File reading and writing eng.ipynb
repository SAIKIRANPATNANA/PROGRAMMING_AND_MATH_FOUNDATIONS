```python
import os
```

```python
class FileHandling:
    def __init__(self):
        pass
    def createFile(self,filename):
        try:
            with open(filename,'w') as f:
                f.write('stuff\n')
        except IOError as e:
            print(e)
    def readFile(self, filename):
        try: 
            with open(filename,'r') as f:
                lines = f.readlines()
                for line in lines:
                      print(line)
        except IOError as e:
            print(e)
    def appendFile(self, filename, text):
        try:
            with open(filename, 'a') as f:
                f.write(text)
        except IOError as e:
            print(e)
    def renameFile(self, filename, newfilename):
        try: 
            os.rename(filename, newfilename)
        except IOError as e:
            print(e) 
    def deleteFile(self, filename):
        try: 
            os.remove(filename)
        except IOError as e:
            print(e)
```

```python
filehandle = FileHandling()
```

```python
filename = "example.txt"
newfilename = "new_example.txt"
filehandle.createFile(filename)
filehandle.readFile(filename)
filehandle.appendFile(filename, "This is some additional text.\n")
filehandle.readFile(filename)
filehandle.renameFile(filename, newfilename)
filehandle.readFile(newfilename)
filehandle.deleteFile(newfilename)
filehandle.readFile(newfilename)
```
